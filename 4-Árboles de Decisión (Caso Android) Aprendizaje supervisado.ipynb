# Árboles de decisión

En este caso de uso práctico se pretende resolver un problema de detección de malware en dispositivos Android mediante el análisis del tráfico de red que genera el dispositivo mediante el uso de árboles de decisión.

## Conjunto de datos: Detección de malware en Android

### Descripción
El malware avanzado y sofisticado para Android es capaz de identificar la presencia de un emulador utilizado por el analista de malware y, en respuesta, alterar su comportamiento para evadir la detección. Para superar este problema, instalamos las aplicaciones Android en un dispositivo real y capturamos su tráfico de red. Consulta nuestra Sandbox de Android disponible públicamente.

El conjunto de datos **CICAAGM** se capturó instalando las aplicaciones Android en teléfonos inteligentes reales de forma semi-automatizada. El conjunto de datos fue generado a partir de **1900 aplicaciones** divididas en las siguientes tres categorías:

**1. Adware (250 aplicaciones)**
* **Airpush**: Diseñado para enviar anuncios no solicitados al sistema del usuario y robar información.
* **Dowgin**: Diseñado como una biblioteca de anuncios que también puede robar información del usuario.
* **Kemoge**: Diseñado para tomar el control del dispositivo Android del usuario. Este adware es un híbrido de botnet y se disfraza como aplicaciones populares mediante reempaquetado.
* **Mobidash**: Diseñado para mostrar anuncios y comprometer la información personal del usuario.
* **Shuanet**: Similar a Kemoge, también está diseñado para tomar el control del dispositivo del usuario.

**2. Malware general (150 aplicaciones)**
* **AVpass**: Diseñado para distribuirse disfrazado como una aplicación de reloj.
* **FakeAV**: Diseñado como una estafa que engaña al usuario para que compre una versión completa del software para solucionar infecciones inexistentes.
* **FakeFlash/FakePlayer**: Diseñado como una aplicación Flash falsa para redirigir a los usuarios a un sitio web (después de instalarse con éxito).
* **GGtracker**: Diseñado para fraude por SMS (envía mensajes SMS a números de tarifa premium) y robo de información.
* **Penetho**: Diseñado como un servicio falso (herramienta de hackeo para dispositivos Android que puede utilizarse para descifrar contraseñas WiFi). El malware también puede infectar la computadora del usuario a través de archivos adjuntos infectados, actualizaciones falsas, medios externos y documentos infectados.

**3. Benignas (1500 aplicaciones)**
* Aplicaciones del mercado de GooglePlay de 2015 (más populares gratuitas y nuevas gratuitas)
* Aplicaciones del mercado de GooglePlay de 2016 (más populares gratuitas y nuevas gratuitas)

## Funciones auxiliares

```python
# Construcción de una función que realice el particionado completo
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    # Divide el DataFrame en conjuntos de entrenamiento (60%), validación (20%) y prueba (20%)
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
# Separa el DataFrame en variables predictoras (X) y variable objetivo (y)
def remove_labels(df, label_name):
    X = df.drop(label_name, axis=1)
    y = df[label_name].copy()
    return (X, y)
```

```python
# Evalúa y compara el desempeño del modelo con y sin preprocesamiento usando una métrica dada
# coparar si los datos normales tienen mejor predicción en los árboles que los datos escalables
def evaluate_result(y_pred, y, y_prep_pred, y_prep, metric):
    print(metric.__name__, "WITHOUT preparation:", metric(y_pred, y, average='weighted'))
    print(metric.__name__, "WITH preparation:", metric(y_prep_pred, y_prep, average='weighted'))
```

## 3. División del conjunto de datos

```python
# Dividimos el conjunto de datos
train_set, val_set, test_set = train_val_test_split(X)
```

```python
X_train, y_train = remove_labels(train_set, 'calss')
X_val, y_val = remove_labels(val_set, 'calss')
X_test, y_test = remove_labels(test_set, 'calss')
```

## 4. Escalando el conjunto de datos

Es importante comprender que los árboles de decisión son algoritmos que **no requieren demasiada preparación de los datos** concretamente, no requieren la realización de escalado o normalización. Se va a demostrar como aplicar preprocesamientos como el escalado y como puede incluso llegar a afectar al rendimiento del modelo.

```python
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_val_scaled = scaler.fit_transform(X_val)
```

```python
# Transformación a un DataFrame de Pandas
X_train_scaled = DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_train_scaled.head(10)
```

```python
X_train_scaled.describe()
```

## 5. Árbol de decisión

```python
from sklearn.tree import DecisionTreeClassifier

MAX_DEPTH = 20

# Modelo entrenado con el conjunto de datos sin escalar
clf_tree = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=42)
clf_tree.fit(X_train,y_train)
```

```python
# Modelo entrenado con el conjunto de datos escalado
clf_tree_scaled = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=42)
clf_tree_scaled.fit(X_train_scaled, y_train)
```

Comenzar prediciendo con el propio conjunto de datos con el que se ha entrenado el algoritmo (train set), suele ser interesante para comprobar si se esta produciendo overfitting.

```python
# Predecimos con el conjunto de datos de entrenamiento
y_train_pred = clf_tree.predict(X_train)
y_train_prep_pred = clf_tree_scaled.predict(X_train_scaled)
```

```python
# Comparamos resultados entre escalado y sin escalar
evaluate_result(y_train_pred, y_train, y_train_prep_pred, y_train, f1_score)
```

```python
# Probando los datos sin preparar

from sklearn.metrics import precision_score, recall_score, f1_score

y_pred_tree = clf_tree.predict(X_test)  #  clf_tree is your trained DecisionTreeClassifier

print("Árbol de Decisión - SIN preparación de datos")
precision = precision_score(y_test, y_pred_tree, average="weighted")
recall = recall_score(y_test, y_pred_tree, average="weighted")
f1 = f1_score(y_test, y_pred_tree, average="weighted")

print(f"Precisión: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
# Predecimos con el conjunto de datos de validación
y_pred = clf_tree.predict(X_val)
y_prep_pred = clf_tree_scaled.predict(X_val_scaled)
```

```python
# Comparamos resultados entre escalado y sin escalar
evaluate_result(y_pred, y_val, y_prep_pred, y_val, f1_score)
```

```python
# Probando los datos escalados
y_pred_tree_scaled = clf_tree_scaled.predict(X_test_scaled)

print("Árbol de Decisión - CON preparación de datos")
precision = precision_score(y_test, y_pred_tree_scaled, average="weighted")
recall = recall_score(y_test, y_pred_tree_scaled, average="weighted")
f1 = f1_score(y_test, y_pred_tree_scaled, average="weighted")

print(f"Precisión: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

Aunque los resultados son muy similares, el modelo sin escalar es más rápido y proporciona resultados ligeramente superiores. **A partir de este punto, y, en función de los resultados que hemos visto, nos quedamos con la opción de no escalar nuestros conjuntos de datos**

## 6. Visualizando el limite de decision

```python
# Reducimos el número de atributos del conjunto de datos para visualizarlo mejor
# colocamos los campos con los que se va a entrenar
X_train_reduced = X_train[['min_flowpktl','flow_fin']]
```

```python
# Generamos un modelo con el conjunto de datos reducido
clf_tree_reduced = DecisionTreeClassifier(max_depth=10, random_state=42)
clf_tree_reduced.fit(X_train_reduced, y_train)
```

```python
# Representamos gráficamente el límite de decisión construido
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
%matplotlib inline

def plot_decision_boundary(clf, X, y, plot_training=True, resolution=1000):
    mins = X.min(axis=0) - 1
    maxs = X.max(axis=0) + 1
    x1, x2 = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
    plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="normal") # amarillo
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="adware") # azul
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="malware") # verde
        plt.axis([mins[0], maxs[0], mins[1], maxs[1]])
    plt.xlabel('min_flowpktl', fontsize=14)
    plt.ylabel('flow_fin', fontsize=14, rotation=90)

plt.figure(figsize=(12, 6))
plot_decision_boundary(clf_tree_reduced, X_train_reduced.values, y_train)
plt.show()
```
