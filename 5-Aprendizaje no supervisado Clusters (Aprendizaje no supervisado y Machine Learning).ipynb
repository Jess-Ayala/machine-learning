### Descripción
El conjunto de datos contiene transacciones realizadas con tarjetas de crédito en septiembre de 2013 por titulares de tarjetas europeos. Este conjunto de datos presenta transacciones que ocurrieron durante dos días, en los cuales se registraron 492 fraudes de un total de 284,807 transacciones. El conjunto de datos está altamente desbalanceado; la clase positiva (fraudes) representa solo el 0.172% de todas las transacciones.

Contiene únicamente variables de entrada numéricas, las cuales son el resultado de una transformación mediante Análisis de Componentes Principales (PCA). Desafortunadamente, debido a cuestiones de confidencialidad, no podemos proporcionar las características originales ni más información de contexto sobre los datos. Las características V1, V2, … V28 son los componentes principales obtenidos mediante PCA. Las únicas características que no fueron transformadas mediante PCA son ‘Time’ y ‘Amount’. La característica ‘Time’ contiene los segundos transcurridos entre cada transacción y la primera transacción del conjunto de datos. La característica ‘Amount’ representa el monto de la transacción; esta característica puede usarse, por ejemplo, para aprendizaje sensible al costo dependiente del ejemplo. La característica ‘Class’ es la variable de respuesta y toma el valor 1 en caso de fraude y 0 en caso contrario.

## Funciones auxiliares

```python
# crear una grafica x e y
def plot_data(X, y):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], 'k.', markersize=2)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], 'r.', markersize=2)
# grafica para visualizar los centroides
def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=30, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=50, linewidths=50,
                color=cross_color, zorder=11, alpha=1)
# marcar con ejes y colores para los closters
def plot_decision_boundaries(clusterer, X, y, resolution=1000, show_centroids=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X, y)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)
```

```python
def purity_score(y_true, y_pred):
    # Calcula la matriz de contingencia (también conocida como matriz de confusión)
    # Esta matriz compara las etiquetas verdaderas (y_true) con las etiquetas predichas (y_pred)
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)

    # Calcula y devuelve el puntaje de pureza
    # np.amax(..., axis=0): toma el valor máximo de cada columna (es decir, el número de muestras más frecuentes en cada cluster)
    # np.sum(...) suma todos esos máximos, representando la cantidad total de clasificaciones correctas dentro de cada cluster
    # Se divide entre el total de elementos en la matriz para normalizar (obtener una proporción)
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)
```

Otra opción, es eliminar las características que no se encuentran en un rango similar siempre y cuando no sean muy influyentes para la predicción.

```python
df = df.drop(["Time","Amount"], axis=1)
```

## 4. Kmeans con un conjunto de datos de dos dimensiones

Antes de comenzar con el entrenamiento de KMEANS para todos los atributos del conjunto de datos, se realiza una prueba para dos atributos con el objetivo de comprender como construye el límite de decisión.

```python
X = df[["V10","V14"]].copy()
X
```

```python
# Generamos los clusters para nuestro conjunto de datos sin etiquetar
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(12, 6))
plot_decision_boundaries(kmeans, X.values, df["Class"].values)
plt.xlabel(..., fontsize=14)
plt.ylabel(..., fontsize=14)
plt.show()
```

```python
# Se cuenta cuántas veces aparece cada etiqueta en la lista de clusters (convertida a lista con .tolist())
# Esto da el total de muestras por cada grupo (cluster)
counter = Counter(clusters.tolist())

# Se cuentan los casos en que 'Class' es 1 (maliciosos/fraudulentos) dentro de cada cluster
# Primero se filtran las filas donde Class == 1, luego se obtiene la lista de sus etiquetas de cluster
bad_counter = Counter(clusters[df['Class'] == 1].tolist())

# Se recorre cada clave (etiqueta de cluster) ordenada
for key in sorted(counter.keys()):
    # Se imprime cuántas muestras hay en total para esa etiqueta y cuántas son maliciosas
    # Si no hay maliciosas en ese cluster, bad_counter[key] devolverá 0 automáticamente
    print("Label {0} has {1} samples - {2} are malicious samples".format(
        key, counter[key], bad_counter[key]))
```

## 5. Kmeans con un conjunto de datos multidimensional

Al tratarse de un algoritmo basado en aprendizaje no supervisado, no se requieren las etiquetas para entrenar o predir. Hay que tener en cuenta que en este tipo de problemas se presupone que no se dispone de las etiquetas.

```python
X = df.drop("Class",axis=1)
y = df["Class"].copy()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
# Evaluamos los clusters y el contenido que se han formado

# Se cuenta cuántas veces aparece cada etiqueta (cluster) en la lista de predicciones de clusters
# Esto representa cuántas muestras hay en cada cluster en total
counter = Counter(clusters.tolist())

# Se filtran los elementos en 'clusters' cuya clase real (y) es 1 (maliciosa) y se cuentan
# Esto da el número de muestras maliciosas en cada cluster
bad_counter = Counter(clusters[y == 1].tolist())

# Se recorre cada etiqueta de cluster en orden ascendente
for key in sorted(counter.keys()):
    # Se imprime cuántas muestras hay en total y cuántas son maliciosas para ese cluster
    # bad_counter[key] devolverá 0 si no hay ninguna muestra maliciosa en ese cluster
    print("Label {0} has {1} samples - {2} are malicious samples".format(
        key, counter[key], bad_counter[key]))
```

## 6. Reducción del número de características

En muchas ocasiones Kmeans funciona mejor con un número de características bajo.

### Aplicamos selección de características con _Random Forest_

```python
# Utilizamos Random Forest para realizar selección de características
#Tarda aprox 5 minutos
from sklearn.ensemble import RandomForestClassifier

clf_rnd = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
clf_rnd.fit(X, y)
```

```python
# Seleccionamos las características más importantes

# Se crea un diccionario que asocia cada nombre de característica (columna del DataFrame df)
# con su importancia correspondiente obtenida del clasificador clf_rnd
# clf_rnd.feature_importances_ es un arreglo con un valor por cada característica
feature_importances = {name: score for name, score in zip(list(df), clf_rnd.feature_importances_)}

# Se convierte el diccionario anterior en un objeto Series de pandas
# Luego se ordenan las importancias de forma descendente (de mayor a menor importancia)
feature_importances_sorted = pd.Series(feature_importances).sort_values(ascending=False)
```

```python
# Reducimos el conjunto de datos a las 7 características más importantes
X_reduced = X[list(feature_importances_sorted.head(7).index)].copy()
```

```python
X_reduced
```

### Entrenamiento de KMEANS con el conjunto de datos reducido

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_reduced)
```

```python
# Evaluamos los clusters y el contenido que se han formado
counter = Counter(clusters.tolist())
bad_counter = Counter(clusters[y == 1].tolist())

for key in sorted(counter.keys()):
    print("Label {0} has {1} samples - {2} are malicious samples".format(
        key, counter[key], bad_counter[key]))
```

## 7. Evaluación de los resultados

Hay que tener cuidado con los conjuntos de datos desequilibrados cuando se utilizan métricas para medir la pureza de los clusters. Una posible solución es utilizar técnicas de equilibrado del conjunto de datos, como la generación de más ejemplos de transacciones fraudulentas o la disminución de ejemplos de transacciones legítimas.

```python
# Calculamos el purity score, es importante darse cuenta de que recibe las etiquetas
print("Purity Score:", purity_score(y, clusters))
```

```python
# Calculamos el coeficiente de Shiloutte, es importante darse cuenta de que no le pasamos las etiquetas
# Evalúa cada punto individualmente, considerando qué tan bien está asignado a su cluster en comparación con otros clusters.

# Silhouette Score cercano a 1.0 → excelente separación entre clusters.
# Cerca de 0 → clusters solapados, difícil de distinguir.
# Negativo → muchos puntos están en el cluster incorrecto.

print("Shiloutte: ", metrics.silhouette_score(X_reduced, clusters, sample_size=10000))
```
