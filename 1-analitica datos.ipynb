### Análisis de Datos con Python

**Cargando un conjunto de datos**

Usaremos el conjunto de datos `titanic` de la librería `seaborn`, que contiene información sobre los pasajeros del Titanic.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # permite generar graficas

# Cargar el conjunto de datos Titanic
# en 'data' cargamos todos los datos del titanic
data = sns.load_dataset("titanic")

# Mostrar las primeras filas del dataset
data.head()
```


### Gráfico de barras: Supervivencia por género

```python
 # llamaos a la libreria | cuentame y muestra en | hue -> contraste entre
sns.countplot(data=data, x='sex', hue='survived')
plt.title('Supervivencia por género')
plt.show()
```

```python
# prompt: convierte el campo de "survived" a entero para la gráfica 
data['survived'] = data['survived'].astype(int)
```

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6)) # Declaramos un objeto | configuramos el tamaño de la grafica

# Crear una cuadrícula de gráficos (FacetGrid) donde cada columna representa una clase del barco ('pclass')
g = sns.FacetGrid(data, col="pclass", height=5) 

# Mapear un gráfico de barras (countplot -> contar) a cada faceta para contar la cantidad de pasajeros por género ('sex')
# Se usa la paleta de colores 'Set2' y se diferencian por la variable 'survived' (supervivencia)
g.map(sns.countplot, "sex", palette='Set2', dodge=True, hue_order=[1, 0], data=data, hue='survived')

# Agregar una leyenda al gráfico con etiquetas personalizadas
g.add_legend(title='Supervivencia', labels=['Sobrevivió', 'No sobrevivió'])

# Ajustar los márgenes superiores para evitar que el título se superponga
plt.subplots_adjust(top=0.9)

# Agregar un título general a la figura
g.fig.suptitle('Supervivencia por género y clase')

# Mostrar el gráfico
plt.show()
```

### Gráfico de cajas (bijotes): Tarifas por clase

```python
# Crear un gráfico de caja (boxplot) para visualizar la distribución de tarifas ('fare') en cada clase de pasajero ('pclass')
sns.boxplot(data=data, x='pclass', y='fare') # y -> lo contrastamos con la tarifa

# Agregar un título al gráfico
plt.title('Distribución de tarifas por clase')

# Mostrar el gráfico
plt.show()
```

### Gráfico de dispersión: Edad vs Tarifa

```python
# Crear un gráfico de dispersión (scatter plot) para analizar la relación entre la edad y la tarifa pagada
plt.scatter(data['age'], data['fare'], alpha=0.5) # alphs -> que tan  tenue van a ser los datos (casí transparente)

plt.title('Edad vs Tarifa')

# Etiquetar el eje X con 'Edad'
plt.xlabel('Edad')

# Etiquetar el eje Y con 'Tarifa'
plt.ylabel('Tarifa')

plt.show()
```

```python
# Configurar el tamaño de la figura para mejorar la visualización
plt.figure(figsize=(10, 6)) # tamaño del grafico

# Crear un gráfico de barras para contar el número de sobrevivientes y fallecidos según la ciudad de embarque
sns.countplot(data=data, x='embark_town', hue='survived', palette='coolwarm') # palette -: color de la paleta

plt.title('Supervivencia por ciudad de embarque')

plt.xlabel('Ciudad de embarque')

plt.ylabel('Número de pasajeros')

# Agregar una leyenda que indique qué representa cada color en términos de supervivencia
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# diferenciando entre los que sobrevivieron y los que no
sns.countplot(data=data, x='deck', hue='survived', palette='viridis')

plt.title('Supervivencia por deck')

plt.xlabel('Deck')

plt.ylabel('Número de pasajeros')

plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])

plt.show()
```

## Supervivencia por cubierta

```python
plt.figure(figsize=(10, 6))

# Crear un gráfico de barras que muestra la cantidad de pasajeros en cada cubierta,
# diferenciando entre los que sobrevivieron y los que no
sns.countplot(data=data, x='deck', hue='survived', palette='viridis')

plt.title('Supervivencia por cubierta')

plt.xlabel('Deck')

plt.ylabel('Número de pasajeros')

plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])

plt.show()
```

## Gráfica de barplot

```python
# Ejemplo con barplot

plt.figure(figsize=(8,5))
sns.barplot(data=data, x='pclass', y='survived', palette='pastel')
plt.title('Número de Pasajeros Supervivientes y No Supervivientes por Clase')
plt.xlabel('Clase')
plt.ylabel('Número de Pasajeros')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])
plt.show()
```

## Gráfica de scatterplot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='fare', hue='survived', alpha=0.5, palette='coolwarm')
plt.title('Edad vs Tarifa (Coloreado por Supervivencia)')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])
plt.show()
```

## Gráfica de violinplot

```python
plt.figure(figsize=(10, 6))
# violinplot -> funcion
sns.violinplot(data=data, x='sex', y='age', hue='survived', split=True, palette='muted')
plt.title('Distribución de Edad por Género y Supervivencia')
plt.xlabel('Género')
plt.ylabel('Edad')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])
plt.show()
```

# Análisis de Datos con el Dataset `tips`

Usaremos el dataset `tips`, que contiene información sobre cuentas y propinas en un restaurante.
Mediante gráficas, analizaremos cómo varían las propinas según diferentes factores.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt # libreria de personas que dan propina

# Cargar el dataset 'tips'
datos = sns.load_dataset('tips')

# Mostrar las primeras filas
datos.head()

# tip -> propina | size -> numero de personas en la mesa
```

```python
# Conoce el tipo de dato de cada una de las columnas
datos.info()
```

## Visualización de Datos

**Gráfico de barras (`barplot`)**: Muestra la propina promedio por día de la semana.

```python
plt.figure(figsize=(8,5))
sns.barplot(data=datos, x='day', y='tip', palette='pastel')
plt.title('Propina promedio de la semana')
plt.xlabel('Día')
plt.ylabel('Propinas')
#plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sobrevivió'])
plt.show()
```

**Gráfico de dispersión (`scatterplot`)**: Relación entre el total de la cuenta y la propina.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datos, x='total_bill', y='tip', hue='sex', alpha=0.5, palette='coolwarm')
plt.title('Total de cuenta VS Propina (Coloreado por género)')
plt.xlabel('Total cuenta')
plt.ylabel('Propina')
plt.legend(title='Sexo', labels=['Male', 'Female'])
plt.show()
```

**Gráfico de violín (`violinplot`)**: Distribución de propinas por género.

```python
# Reto 3. Gráfico de violín (`violinplot`): Distribución de propinas por género.
plt.figure(figsize=(10, 6))
sns.violinplot(data=datos, x='sex', y='tip', hue='smoker', hue_order=['No', 'Yes'], split=True, palette='muted')
plt.title('Propinas por género y hábido de fumar')
plt.xlabel('Género')
plt.ylabel('Propina')
plt.legend(title='Fumador', labels=['Si', 'No'])
plt.show()
```

**Mostrar el horario donde hay mas propinas**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datos, x='day', y='tip', hue='sex', alpha=0.5, palette='coolwarm')
plt.title('Horario que dan más propina')
plt.xlabel('Día')
plt.ylabel('Horario')
plt.legend(title='Sexo', labels=['Male', 'Female'])
plt.show()
```
